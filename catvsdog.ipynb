```python
from keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
import keras
```

```python
train_dir = "./cats-dogs/training_set/training_set/"
test_dir = "./cats-dogs/test_set/test_set/"
```

```python
train_datagen = ImageDataGenerator(rescale=1.0/255,rotation_range=0.4)
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=128,
    class_mode='binary'
)
```

```python
test_datagen = ImageDataGenerator(rescale=1.0/255,rotation_range=0.4)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150,150),
    batch_size=128,
    class_mode='binary'
)
```

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(filters=32,kernel_size=3,strides=(1,1),padding="valid",activation='relu',input_shape=[150,150,3]),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu'),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(units=256,activation ='relu'),
    keras.layers.Dense(units=128,activation ='relu'),
    keras.layers.Dense(units=2,activation ='softmax')

])
from tensorflow.keras.optimizers import RMSprop
model.compile(loss = 'sparse_categorical_crossentropy',
                optimizer = RMSprop(lr=0.001),
                metrics=['accuracy'])
```

```python
model.fit(train_generator,steps_per_epoch=8,epochs=15,validation_data=test_generator,validation_steps=8,verbose=2)
```

```python
from keras.callbacks import ModelCheckpoint
model_path = "/Users/jarvis/pymycod/tensorflow_AI/trained_models//model_catsvsdogs.h5"
checkpoint = ModelCheckpoint(model_path,monitor="val_accuracy",verbose=1,save_best_only=True,mode=max)
callbacks_list=[checkpoint]
```

```python
history = model.fit(train_generator,epochs=100,verbose=2,validation_data=test_generator,callbacks=callbacks_list)
```
